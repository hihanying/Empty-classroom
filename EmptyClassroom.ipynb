```python
from urllib import request
from bs4 import BeautifulSoup
import re

url = 'https://jwxt.bupt.edu.cn/zxqDtKxJas.jsp'

content = request.urlopen(url)
soup = BeautifulSoup(content)

tags1 = soup.find_all('td',attrs={'class':'sortable','valign':'top'})
list1 = []
for tag in tags1:
    str1 = str(tag.string)
    list1.append(re.sub('\s','',str1))

tags2 = soup.find_all('td',attrs={'bgcolor':'white','align':'left'})
list2 = []
for tag in tags2:
    str2 = tag.get_text()
    str2 = re.sub('\s','',str2).replace("教一楼:","").replace("教二楼:","").replace("教三楼:","").replace("教四楼:","")
    if(str2[-1]==","):
        str2 = str2[:-1]
    list2.append(str2)

dict1 = {}
for item in list2:
    crs=item.split(",")
    for cr in crs:
        if cr in dict1.keys():
            dict1[cr].append(list1[list2.index(item)])
        else:
            dict1[cr]=[list1[list2.index(item)]]
dict1
```
